```python
import kagglehub
import pandas as pd
import os # Import the os module to interact with the operating system

# Download latest version
path = kagglehub.dataset_download("rjmanoj/credit-card-customer-churn-prediction")

print("Path to dataset files:", path)

# List the files in the downloaded directory to find the CSV file
files_in_directory = os.listdir(path)
print("Files in the downloaded directory:", files_in_directory)

# Assuming the main dataset file is named 'Churn_Modelling.csv' based on common Kaggle dataset structures
# You might need to adjust this filename based on the actual output of os.listdir(path)
csv_file_name = 'Churn_Modelling.csv'
csv_file_path = os.path.join(path, csv_file_name)

# Read the specific CSV file
df = pd.read_csv(csv_file_path)
df.head()
```

```python
df = df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df.head()
```

```python
df
```

```python
df1=pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df1.drop(['Exited'],axis=1)
y=df1['Exited']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(3,activation='relu',input_dim=11))
# model.add(Dense(3,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy')
```

```python
histry=model.fit(X_train_scaled,y_train,epochs=200)
```

```python
y_log=model.predict(X_test_scaled)
```

```python
y_pred=np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
